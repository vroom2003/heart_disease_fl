```python
z
import pandas as pd
import numpy as np

print("ðŸ§¹ Cleaning Heart Disease Dataset...")

# Load raw
file_path = r'C:\Users\ACER\Desktop\Heart_disease_fl\data\heart_disease.csv'
```

```python
df = pd.read_csv(file_path)
print(df.head())
```

```python
print(df.info())
```

```python
columns = ['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal','target']
df.columns = columns
```

```python
# Replace '?' with NaN
df.replace('?', np.nan, inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
# Boxplot to visually inspect outliers
plt.figure(figsize=(12, 6))
sns.boxplot(data=df.select_dtypes(include=[np.number]))
plt.xticks(rotation=90)  # Rotate column names for readability
plt.show()
```

```python
print(df.isnull().sum())
```

```python
# 1. Check for duplicates
# -----------------------
duplicates = df.duplicated()
print(f"Number of duplicate rows: {duplicates.sum()}")

# -----------------------
# 2. Show duplicate rows (if any)
# -----------------------
if duplicates.sum() > 0:
    print("Duplicate rows:")
    print(df[df.duplicated()])

# -----------------------
# 3. Drop duplicates (if you want)
# -----------------------
df = df.drop_duplicates()
print(f"Shape after removing duplicates: {df.shape}")
```

```python
correlation_matrix = df.select_dtypes(include='number').corr()

# Print the correlation matrix
print(correlation_matrix)
```
